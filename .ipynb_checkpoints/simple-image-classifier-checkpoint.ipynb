# Simple Image Classifier

In this notebook I build a classifier for the Fashion-MNIST dataset - a collection of images of clothing items, split into 10 classes. Each image consists of a 28x28 grid of greyscale pixels.

The data are pre-split into training and testing dataframes, with an 85%/15% split.


https://www.kaggle.com/zalando-research/fashionmnist

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import keras

from sklearn.model_selection import train_test_split
from keras.utils import to_categorical

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Dense, Dropout
from keras.layers import Flatten, BatchNormalization
```

Importing the training and test dataset:

```python
train_df = pd.read_csv('~/resources/fashion-mnist_train.csv')
test_df = pd.read_csv('~/resources/fashion-mnist_test.csv')
train_df.head()
```

The data are contained in a pandas dataframe, where each column contains a pixel value, and each row is a different image.

We first convert the pixel columns to numpy arrays, and the label column to a binary class matrix. We also create a dictionary to interpret the label matrix according to the dataset guidance.

```python
train_data = np.array(train_df.iloc[:, 1:])
test_data = np.array(test_df.iloc[:, 1:])

train_labels = to_categorical(train_df.iloc[:, 0])
test_labels = to_categorical(test_df.iloc[:, 0])

label_dict = {0: 'T-shirt/top',
              1: 'Trouser',
              2: 'Pullover',
              3: 'Dress',
              4: 'Coat',
              5: 'Sandal',
              6: 'Shirt',
              7: 'Sneaker',
              8: 'Bag',
              9: 'Ankle boot'}
```

We reshape the images into easy interpretable grids, and normalise them.

```python
rows, cols = 28, 28

train_data = train_data.reshape(train_data.shape[0], rows, cols, 1)
test_data = test_data.reshape(test_data.shape[0], rows, cols, 1)

train_data = train_data.astype('float32')
test_data = test_data.astype('float32')

train_data /= 255.0
test_data /= 255.0
```

We next want to visualise a subset of the training data and their labels.

```python
n_subplots = 24

fig, axs = plt.subplots(math.ceil(n_subplots/6), 6, figsize=(15, 7), sharey=True)

axs = axs.ravel()
for ax in axs:
    ax.axis('off') 
    
for i in range(n_subplots):
    axs[i].imshow(train_data[i], cmap='gray_r')
    axs[i].set_title(label_dict[train_labels[i].argmax()])
```

We split the training data to create a small 20% validation dataset.

```python
train_x, val_x, train_y, val_y = train_test_split(train_data, train_labels, test_size=0.2)
```

# Creating a Convolutional Neural Network

Building the model

```python
batch_size = 256
epochs = 5
input_shape = (rows, cols, 1)
```

```python
def baseline_model():
    model = Sequential()
    model.add(BatchNormalization(input_shape=input_shape))
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=(2,2)))
    model.add(Dropout(0.25))
    
    model.add(BatchNormalization())
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))

    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(10, activation='softmax'))
    return model
```

```python
model = baseline_model()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
model.summary()
```

Training the model

```python
history = model.fit(train_x, train_y,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(val_x, val_y))
```

```python
hist = history.history
print(hist)
```

Visualising the results of the training. The validation accuracy appears to be levelling.

```python
fig, ax = plt.subplots(2)

ax[0].plot(hist['accuracy'])
ax[0].plot(hist['val_accuracy'])
ax[0].legend(['training accuracy', 'validation accuracy'])

ax[1].plot(hist['loss'])
ax[1].plot(hist['val_loss'])
ax[1].legend(['training loss', 'validation loss'])

for axs in ax.flat:
    axs.label_outer()
```

## Forming predictions on unseen data

```python
predict = model.predict(test_data)
```
